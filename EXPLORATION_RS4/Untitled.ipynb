```python
# os: Operating System ,          (Ex.   ,     )
# urllib: URL      . (Ex. urllib.request, urllib.parse, ...)
# cv2: OpenCV ,       
# numpy(NumPy):          .          
# pixellib:    segmentation   . 
# pixellib.semantic: segmentation  , semantic segmentation      
# matplotlib:       NumPy    ,   
import os
import urllib
import cv2
import numpy as np
import tensorflow as tf
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt

# colab setting
'''
!pip install pixellib
file_path = '/usr/''local/lib/python3.10/dist-packages/pixellib/semantic/deeplab.py'
old_text = 'from tensorflow.python.keras.layers import BatchNormalization'
new_text = 'from keras.layers.normalization.batch_normalization import BatchNormalization'
def replace_in_file(file_path, old_text, new_text):
    with open(file_path, 'r') as file:
        file_content = file.read()

    file_content = file_content.replace(old_text, new_text)

    with open(file_path, 'w') as file:
        file.write(file_content)
replace_in_file(f''ile_path, old_text, new_text)
'''
# os   getenv()      file_path 
#     ,   
# cv2.imread():       
# img_path = '/content/drive/MyDrive/Colab Notebooks/EXPLORATION_data/E05/my_image.png'

def img_segmantation(img_path):    
    img_orig = cv2.imread(img_path) 

    #     print(img_orig.shape)

    # cv2.cvtColor( ,   ):     
    # cv2.COLOR_BGR2RGB:     (BGR  RGB  )
    # plt.imshow():     ,  RGB(A)   2D  
    # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html
    # plt.show():    figure  ( figure ,  )
    # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
    plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
    plt.show()


    #    
    # 1. os.getenv(x)  x     . model_dir  "/aiffel/human_segmentation/models" 
    # 2. #os.path.join(a, b)      model_file  "/aiffel/aiffel/human_segmentation/models/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5" 
    # 1
    # model_dir = '/content/drive/MyDrive/Colab Notebooks/EXPLORATION_data/E05'
    # 2
    # model_file = '/content/drive/MyDrive/Colab Notebooks/EXPLORATION_data/E05/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5'
    model_file = '/aiffel/aiffel/human_segmentation/models/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5'
    if not os.path.exists(model_file):
        # PixelLib   url
        model_url = 'https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5'

        #  
        urllib.request.urlretrieve(model_url, model_file) # urllib    request  urlretrieve   model_url     model_file  

    model = semantic_segmentation() #PixelLib      semantic segmentation    
    model.load_pascalvoc_model(model_file) # pascal voc    (model_file)   

    segvalues, output = model.segmentAsPascalvoc(img_path) # segmentAsPascalvoc()       ,    ,   pacalvoc    


    #pascalvoc  
    LABEL_NAMES = [
        'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
        'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
        'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
    ]
    #     len(LABEL_NAMES)

    #segmentAsPascalvoc()           output matplotlib  
    plt.imshow(output)
    plt.show()

    segvalues # segmentAsPascalvoc()           

    #segvalues  class_ids    pacalvoc   
    for class_id in segvalues['class_ids']:
        print(LABEL_NAMES[class_id])

    #     
    # PixelLib   
    #      !


    #  
    colormap = np.zeros((256, 3), dtype = int)
    ind = np.arange(256, dtype=int)

    for shift in reversed(range(8)):
        for channel in range(3):
            colormap[:, channel] |= ((ind >> channel) & 1) << shift
        ind >>= 3

    colormap[:20] # 20  

    if 15 in segvalues['class_ids'] : #person
        seg_color = (128,128,192) #   - colormap  RGB  output  BGR     
    elif 12 in segvalues['class_ids'] : #dog
        seg_color = (128,0,64)

    # output    seg_color  1(True),  0(False) 
    # seg_color  person       gray 
    # cmap      
    seg_map = np.all(output==seg_color, axis=-1) 
    #     print(seg_map.shape) 
    plt.imshow(seg_map, cmap='gray')
    plt.show()


    #  img_show         color_mask      
    img_show = img_orig.copy()

    # True False   255 0 
    img_mask = seg_map.astype(np.uint8) * 255

    # 255 0   
    color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

    #     
    # 0.6 0.4    .
    img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.4, 0.0)

    plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
    plt.show()

    # (13,13) blurring kernel size 
    #  
    img_orig_blur = cv2.blur(img_orig, (130,130))

    # plt.imshow():     .
    # cv2.cvtColor( ,   ):     
    # cv2.COLOR_BGR2RGB:  BGR   
    #     (BGR  RGB  )   
    plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
    plt.show()

    # cv2.cvtColor( ,   ):     
    # cv2.COLOR_BGR2RGB:  BGR   
    #     (BGR  RGB  ) 
    img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)

    # cv2.bitwise_not():  .  0  255 
    #     255  0.
    img_bg_mask = cv2.bitwise_not(img_mask_color)

    # cv2.bitwise_and()       .
    # 0   bitwise_and   0   
    #  0      0 .     !
    img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
    plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
    plt.show()

    #chroma key
    chroma_key_color = [0, 177, 64]
    img_concat = np.where(img_mask_color==255, img_orig, np.array(chroma_key_color, dtype=np.uint8))
    plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
    plt.show()

    if 15 in segvalues['class_ids'] : #person
        # 
        text = 'finger cut'
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1.2
        thickness = 2
        color = (255, 0, 0)  #   (BGR )

        #    
        text_position = (450, 200)

        #   
        image_with_text = cv2.putText(img_concat, text, text_position, font, font_scale, color, thickness)
        plt.cla()
        plt.imshow(cv2.cvtColor(image_with_text, cv2.COLOR_BGR2RGB))
        plt.show()

    elif 12 in segvalues['class_ids'] : #dog
        text1 = 'merge legs'
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 6
        thickness = 20
        color = (255, 0, 0)  #   (BGR )

        #    
        text_position = (500, 3100)

        #   
        image_with_text1 = cv2.putText(img_concat, text1, text_position, font, font_scale, color, thickness)

        text2 = 'out of leg range'
        text2_position = (2000, 3800)

        #     
        image_with_text2 = cv2.putText(image_with_text1, text2, text2_position, font, font_scale, color, thickness)
        plt.imshow(cv2.cvtColor(image_with_text2, cv2.COLOR_BGR2RGB))
        plt.show()


    # np.where(, , )
    #   255       
    #      .
    img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
    # plt.imshow():     .
    # cv2.cvtColor( ,   ):     
    # cv2.COLOR_BGR2RGB:  BGR    
    #     (BGR  RGB  )
    plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
    plt.show()

    # Q.    ,    ?
    #  : img_mask_color   , img_orig, img_orig_blur  .
    img_concat = np.where(img_mask_color==0, img_orig, img_orig_blur)
    plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
    plt.show()

human_img_path = '/aiffel/aiffel/human_segmentation/images/karina.png'
dog_img_path = '/aiffel/aiffel/human_segmentation/images/IMG_0156.png'

img_segmantation(human_img_path)
# img_segmantation(dog_img_path)
```
