```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras import models, layers
from tensorflow.keras import callbacks
from tensorflow.keras import utils
from keras.datasets import reuters
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
# keras.datasets  boston_housing  .
from keras.datasets import boston_housing
import numpy as np

(train_data, train_labels), (test_data, test_labels) = boston_housing.load_data()

#   .
print(len(train_data))
print(len(test_data))

#  1-1:      .
#       .
pre_train_data = train_data - np.mean(train_data,axis=0)

#      .
#    0  ,  1  .
pre_train_data /= np.std(pre_train_data, axis=0)

#     ,  .
pre_test_data = test_data - np.mean(test_data,axis=0)
pre_test_data /= np.std(pre_test_data, axis=0)

#  1-2:         .
# input_shape (train_data.shape[1], ) .
# ()   loss mse, metrics mae .
model = models.Sequential()
model.add(layers.Input(shape=(pre_train_data.shape[1],), name='input'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1))
model.summary()


model.compile(optimizer='adam',
              loss='mse',
              metrics=['mae'])

#  1-3:     .
history = model.fit(pre_test_data, test_labels, 
                     epochs=200,
                     validation_data=(pre_test_data, test_labels)
                     )

#    loss, val_loss, mae, val_mae  .
plt.style.use('seaborn-white')

history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, 'b-', label='train_loss')
ax1.plot(epochs, val_loss, 'r-', label='val_loss')
ax1.set_title('Train and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()

mae = history_dict['mae']
val_mae = history_dict['val_mae']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, mae, 'b-', label='train_mae')
ax2.plot(epochs, val_mae, 'r-', label='val_mae')
ax2.set_title('Train and Validation MAE')

ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.grid()
ax2.legend()

plt.show()

#  1-4:     .
model.evaluate(pre_test_data, test_labels)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from tensorflow.keras import models, layers
from tensorflow.keras import callbacks
from tensorflow.keras import utils
from keras.datasets import reuters

#  2-1: reuters  ,   25%   .
(train_data, train_labels), (test_data, test_labels) = reuters.load_data(num_words=10000)

train_data, val_data, train_labels, val_labels = train_test_split(
    train_data,train_labels, test_size=0.25, random_state=25)

# , ,       .
print(len(train_data))
print(len(val_data))
print(len(test_data))
print(len(set(train_labels)))

#  2-2:       -  .
def one_hot_encoding(data, dim=10000):
  results = np.zeros((len(data), dim))
  for i, d in enumerate(data):
    results[i,d] = 1.
  return results

x_train = one_hot_encoding(train_data)
x_val = one_hot_encoding(val_data)
x_test = one_hot_encoding(test_data)

#     to_categorical()   .
y_train = utils.to_categorical(train_labels)
y_val = utils.to_categorical(val_labels)
y_test = utils.to_categorical(test_labels)

#  2-3:      BatchNormalization Dropout    .
inputs = layers.Input(shape=(10000,))
dense1 = layers.Dense(128,activation='relu')(inputs)
bn1 = layers.BatchNormalization()(dense1) #batchnormal
actv1 = layers.Activation('relu')(bn1) #act
do1 = layers.Dropout(0.5)(actv1) #dropout
dense2 = layers.Dense(128,activation='relu')(do1)
bn2 = layers.BatchNormalization()(dense2) #batchnormal
actv2 = layers.Activation('relu')(bn2) #act
do2 = layers.Dropout(0.5)(actv2) #dropout
dense3 = layers.Dense(46,activation='softmax')(do2)

model = models.Model(inputs=inputs, outputs=dense3)

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

utils.plot_model(model, show_shapes=True)

model.summary()

#  2-4: ModelCheckpoint EarlyStopping      .
check_point_cb = callbacks.ModelCheckpoint('keras_mnist_model.h5', save_best_only=True)
early_stopping_cb = callbacks.EarlyStopping(patience=3, monitor='val_loss',
                                  restore_best_weights=True)
history = model.fit(x_train, y_train, epochs=50,
                    validation_data=(x_val, y_val),
                    callbacks=[check_point_cb, early_stopping_cb])

#   `loss`, `val_loss`, `accuracy`, `val_accuracy`  .
plt.style.use('seaborn-white')

history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, 'b-', label='train_loss')
ax1.plot(epochs, val_loss, 'r-', label='val_loss')
ax1.set_title('Train and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()

accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, accuracy, 'b-', label='train_accuracy')
ax2.plot(epochs, val_accuracy, 'r-', label='val_accuracy')
ax2.set_title('Train and Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.grid()
ax2.legend()

plt.show()

#  2-5:     .
model.evaluate(x_test, y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from tensorflow.keras import models, layers
from tensorflow.keras.datasets import cifar10

#  3-1: cifar10  ,   20%   .
(x_train_full, y_train_full), (x_test, y_test) = cifar10.load_data()

x_train, x_val, y_train, y_val = train_test_split(
    x_train_full,y_train_full, test_size=0.2, random_state=25)


# , ,      .
print("  : {} : {}".format(x_train_full.shape, y_train_full.shape))
print(" : {} : {}".format(x_train.shape, y_train.shape))
print(" : {} : {}".format(x_val.shape, y_val.shape))
print(" : {} : {}".format(x_test.shape, y_test.shape))

# cifar10   'airplane', 'automobile', 'bird', 'cat', 'deer', 
# 'dog', 'frog', 'horse', 'ship', 'truck' class_name .
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)               

#   0 x_train[0]  .
plt.style.use('seaborn-white')

plt.figure()
plt.imshow(x_train[0])
plt.colorbar()
plt.show()

# , ,   (shape) .
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)

print(y_train.shape)
print(y_val.shape)
print(y_test.shape)

#  3-2: , ,   (shape) 32 * 32 * 3 = 3072 .
x_train = x_train.reshape(-1,32*32*3)
x_val = x_val.reshape(-1,32*32*3)
x_test = x_test.reshape(-1,32*32*3)

print(x_train.shape)
print(x_val.shape)
print(x_test.shape)

# , ,   255  0~1   .
x_train = x_train / 255.
x_val = x_val / 255.
x_test = x_test / 255.

#  3-3: BatchNormalization Dropout     ,
# 10     .
inputs = layers.Input(shape=(3072,))
dense1 = layers.Dense(2048,activation='relu')(inputs)
bn1 = layers.BatchNormalization()(dense1) #batchnormal
actv1 = layers.Activation('relu')(bn1) #act
do1 = layers.Dropout(0.5)(actv1) #dropout
dense2 = layers.Dense(1024,activation='relu')(do1)
bn2 = layers.BatchNormalization()(dense2) #batchnormal
actv2 = layers.Activation('relu')(bn2) #act
do2 = layers.Dropout(0.5)(actv2) #dropout
dense3 = layers.Dense(512,activation='relu')(do2)
bn3 = layers.BatchNormalization()(dense3) #batchnormal
actv3 = layers.Activation('relu')(bn3) #act
do3 = layers.Dropout(0.5)(actv3) #dropout
dense4 = layers.Dense(10,activation='softmax')(do3)

model = models.Model(inputs=inputs, outputs=dense4)

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

utils.plot_model(model, show_shapes=True)

#  3-4: ModelCheckpoint EarlyStopping      .
check_point_cb = callbacks.ModelCheckpoint('keras_mnist_model.h5', save_best_only=True)
early_stopping_cb = callbacks.EarlyStopping(patience=3, monitor='val_loss',
                                  restore_best_weights=True)
history = model.fit(x_train, y_train, epochs=50,
                    validation_data=(x_val, y_val),
                    callbacks=[check_point_cb, early_stopping_cb])

#  3-5:   `loss`, `val_loss`, `accuracy`, `val_accuracy`  .
plt.style.use('seaborn-white')

history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, 'b-', label='train_loss')
ax1.plot(epochs, val_loss, 'r-', label='val_loss')
ax1.set_title('Train and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()

accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, accuracy, 'b-', label='train_accuracy')
ax2.plot(epochs, val_accuracy, 'r-', label='val_accuracy')
ax2.set_title('Train and Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.grid()
ax2.legend()

plt.show()

#     .
model.evaluate(x_test, y_test)
```
